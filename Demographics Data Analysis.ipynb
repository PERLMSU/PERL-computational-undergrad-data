```python
# look at when students are entering MSU
pd.value_counts(end_fs11_df['Entering term code'].values, sort=False)
```

```python
#find number of students that graduate, number that do not graduate
total_rows = len(end_fs11_df['grad major code'])
num_empty_rows = pd.isnull(end_fs11_df['Entering Term Major code']).sum()
num_students = total_rows - num_empty_rows
num_non_graduates = pd.isnull(end_fs11_df['grad major code']).sum() - num_empty_rows
num_graduates = num_students - num_non_graduates
perc_grad = num_graduates/num_students*100
print('Total Number of Students:', num_students)
print('Number of Non-Graduates:', num_non_graduates)
print('Number of Graduates:', num_graduates)
print('Percentage of students that Graduate:', perc_grad)
```

```python
# define new df for graduated students only
df_grad = end_fs11_df[end_fs11_df['grad major name'] != ' ']
df_grad = df_grad.reset_index(drop=True)
df_grad
```

```python
# looking at number of double majors
df_double_majors = df_grad[df_grad['other awards'].str.contains('(BS)') | df_grad['other awards'].str.contains('(BA)')]
df_double_majors = df_double_majors.reset_index(drop=True)
df_double_majors
```

```python
num_double_major_graduates = len(df_double_majors['other awards'])
print('Number of double major graduates:', num_double_major_graduates)
perc_double_major = num_double_major_graduates/num_graduates*100
print('Percent of graduates that have double major:', perc_double_major)
```

```python
#look at how students changed/did not change majors
num_no_change = 0
for i in range(len(end_fs11_df['Entering Term Major name'])):
    if end_fs11_df['Entering Term Major name'][i] == end_fs11_df['grad major name'][i]:
        num_no_change += 1
```

```python
# find number/percent of students that did not change majors
print('Number of Students that did not change major:', num_no_change)
perc_no_change = num_no_change/num_graduates*100
print('Percentage of students that did not change major:', perc_no_change)
```

```python
# find number/percent of students that DID change majors
num_change = num_graduates - num_no_change
print('Number of Students that did change major:', num_change)
perc_change = num_change / num_graduates * 100
print('Percentage of students that did change major:', perc_change)
```

```python
#look at students that changed majors
df_change = df_grad[df_grad['Entering Term Major name'] != df_grad['grad major name']]
df_change = df_change.reset_index(drop=True)
df_change
```

```python
# find number of students that did change colleges when changing majors
num_change_college = 0
for i in range(len(df_change['Entering Term College'])):
    if df_change['Entering Term College'][i] != df_change['Grad College'][i]:
        num_change_college += 1
print('Number of students that changed colleges while changing majors:', num_change_college)
perc_change_college = num_change_college / num_change * 100
print('Percentage of students that changed colleges while changing majors:', perc_change_college)
```

```python
# find number of students that did not change colleges when changing majors
num_same_college = 0
for i in range(len(df_change['Entering Term College'])):
    if df_change['Entering Term College'][i] == df_change['Grad College'][i]:
        num_same_college += 1
print('Number of students that stayed in same college when changing majors:', num_same_college)
perc_same_college = num_same_college / num_change * 100
print('Percentage of students that stayed in same college when changing majors:', perc_same_college)
```

```python
cols = ('previous_course', '183', '193H', '191', '184', '294H', '192', '215', '321', '410', '431', '440', '451', '471', '481', '491', '492', '493')
courses = ['183', '193H', '191', '184', '294H', '192', '215', '321', '410', '431', '440', '451', '471', '481', '491', '492', '493']
df_course_progression = pd.DataFrame(columns=cols)
for i in range(len(courses)):
    df_course_progression.set_value(i, 'previous_course', courses[i])
df_course_progression
```

```python
def crse_progress(input_df, crse_one, crse_two, dictionary):
    count = 0 
    for student in dictionary:
        crse_list = dictionary[student]
        for i in range(len(crse_list)):
            if crse_list[i] == crse_one:
                try:
                    if crse_list[i+1] == crse_two:
                        count += 1
                except:
                    continue
        
    index_loc = int(input_df.index.get_indexer_for((input_df[input_df['previous_course'] == crse_one].index)))
    input_df.set_value(index_loc, crse_two, count)
    return(input_df)
```

```python
crse_progress(df_course_progression, '183', '184', new_dict)
```

```python
for i in range(len(df_course_progression)):
    first_course = df_course_progression['previous_course'][i]
    for j in courses:
        second_course = j
        df_course_progression = crse_progress(df_course_progression, first_course, second_course, new_dict)
        
df_course_progression
```

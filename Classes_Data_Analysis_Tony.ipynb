```python
# have function to sort main dataframes by course
def new_class_df(main_df_name, subj_code, crse_code):
    new_df = main_df_name[(main_df_name['subj_code'] == subj_code)&(main_df_name['crse_code'] == crse_code)]
    new_df = new_df.reset_index(drop=True)
    return(new_df)
```

```python
# have function to make bar plot of students from a dataframe
def plot_bar_df(df, ylim, cutoff, title):
    srs = pd.value_counts(df['major_name'].values, sort=True)
    srs_lim = srs[srs > cutoff]
    
    plt.figure(figsize=(8,8))

    ax = srs_lim.plot(kind='bar')
    plt.ylim([0,ylim])
    plt.xlabel('Major', fontsize=14)
    plt.ylabel('Number of Students', fontsize=14)
    plt.title(title, fontsize=15)

    rects = ax.patches

    for rect, label in zip(rects, srs_lim):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2, height, label, ha='center', va='bottom')
```

```python
import pandas as pd
```

```python
df_classes_1 = pd.read_excel('Classes_FS01-US16_1.xlsx')
df_classes_1
```

```python
df_classes_2 = pd.read_excel('Classes_FS01-US16_2.xlsx')
df_classes_2
```

```python
df_classes_all = pd.concat([df_classes_1, df_classes_2], ignore_index=True)
df_classes_all
```

```python
df_classes_phy = df_classes_all[df_classes_all['subj_code'] == 'PHY']
df_classes_phy = df_classes_phy.reset_index(drop=True)
df_classes_phy
```

```python
ls_phy_course = []
for i in range(len(df_classes_phy)):
    if df_classes_phy['crse_code'][i] in ls_phy_course:
        continue
    else:
        ls_phy_course.append(df_classes_phy['crse_code'][i])
```

```python
ls_phy_course.sort()
ls_phy_course
```

```python
#remove 183A, 183B, change to 183, do for all courses like this
```

```python
for i in range(len(df_classes_phy)):
    crse = df_classes_phy['crse_code'][i]
    if crse == '183A':
        df_classes_phy.set_value(i, 'crse_code', '183')
    if crse == '183B':
        df_classes_phy.set_value(i, 'crse_code', '183')
    if crse == '184A':
        df_classes_phy.set_value(i, 'crse_code', '184')
    if crse == '184B':
        df_classes_phy.set_value(i, 'crse_code', '184')
    if crse == '215B':
        df_classes_phy.set_value(i, 'crse_code', '215')
```

```python
#limit to required physics courses
df_classes_phy_req = df_classes_phy[(df_classes_phy['crse_code'] == '183')|(df_classes_phy['crse_code'] == '184')|(df_classes_phy['crse_code'] == '193H')|(df_classes_phy['crse_code'] == '294H')|(df_classes_phy['crse_code'] == '191')|(df_classes_phy['crse_code'] == '192')|(df_classes_phy['crse_code'] == '215')|(df_classes_phy['crse_code'] == '321')|(df_classes_phy['crse_code'] == '410')|(df_classes_phy['crse_code'] == '431')|(df_classes_phy['crse_code'] == '440')|(df_classes_phy['crse_code'] == '451')|(df_classes_phy['crse_code'] == '471')|(df_classes_phy['crse_code'] == '481')|(df_classes_phy['crse_code'] == '491')|(df_classes_phy['crse_code'] == '492')|(df_classes_phy['crse_code'] == '493')]
df_classes_phy_req = df_classes_phy_req.reset_index(drop=True)
df_classes_phy_req
```

```python
df_crse_phy_req_lim = df_classes_phy_req.drop('term_seq_id', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('subj_code', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('sctn_code', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('grade_code', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('repeat_status_flag', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('major_code', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('major_name', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('dept_name', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('college_name', 1)
df_crse_phy_req_lim = df_crse_phy_req_lim.drop('major_type', 1)
df_crse_phy_req_lim
```
